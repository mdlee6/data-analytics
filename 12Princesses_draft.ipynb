```python
import nltk

from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TweetTokenizer
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from string import punctuation
```

```python
clean_princess = []

with open(r"12dancingprincesses.txt") as f:
    for line in f:
        lline = line.lower()
        cline = lline.strip()         #get rid of newline character

        if cline == '': 
            pass               #this will skip over lines that only had a newline and are now blank
        else:
            tknls = word_tokenize(cline)
            
        for token in tknls:
            clean_princess.append(token)
```
