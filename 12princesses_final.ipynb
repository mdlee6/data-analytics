```python
import nltk

from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TweetTokenizer
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from string import punctuation
```

```python
clean_princess = []

with open(r"12dancingprincesses.txt") as f:
    for line in f:
        lline = line.lower()
        cline = lline.strip()         #get rid of newline character

        if cline == '': 
            pass               #this will skip over lines that only had a newline and are now blank
        else:
            tknls = word_tokenize(cline)
            
        for token in tknls:
            clean_princess.append(token)
```

```python
fd_cp = FreqDist(clean_princess)
fd_cp
```

```python
fd_cp.most_common(10)
```

```python
for cp in clean_princess:
    if cp in punctuation:
        clean_princess.remove(cp)
```

```python
len(clean_princess)
```

```python
FreqDist(clean_princess).most_common(10)
```

```python
for x in clean_princess: 
    if x == '’':
        clean_princess.remove(x)
    if x == '‘': 
        clean_princess.remove(x)
```

```python
FreqDist(clean_princess).most_common(10)
```

```python
eng_stopwords = stopwords.words('english')
```

```python
rm_count = 0    #removed
new_princess = []  #list to hold new words

for p in clean_princess:
    if p not in eng_stopwords:
        new_princess.append(p)
    else: rm_count += 1
```

```python
prince_nw = FreqDist(new_princess)
prince_nw.most_common(10)
```
